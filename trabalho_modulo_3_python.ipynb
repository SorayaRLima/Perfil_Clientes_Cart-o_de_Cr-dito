### 2. Limpeza e Preparo dos Dados

````python
# Converter colunas num√©ricas que est√£o como strings
def converter_valor(valor):
    if isinstance(valor, str):
        return float(valor.replace('.', '').replace(',', '.'))
    return valor

df['limite_credito'] = df['limite_credito'].apply(converter_valor)
df['valor_transacoes_12m'] = df['valor_transacoes_12m'].apply(converter_valor)
````

````python
# Verificar tipos de dados ap√≥s convers√£o
df[['limite_credito', 'valor_transacoes_12m']].dtypes
````

````python
# Tratar valores nulos (se necess√°rio)
df.fillna({'escolaridade': 'n√£o informado', 'estado_civil': 'n√£o informado', 'salario_anual': 'n√£o informado'}, inplace=True)
````

````python
# Criar categorias ordenadas para escolaridade
ordem_escolaridade = ['sem educacao formal', 'ensino medio', 'graduacao', 'mestrado', 'doutorado', 'na', 'n√£o informado']
df['escolaridade'] = pd.Categorical(df['escolaridade'], categories=ordem_escolaridade, ordered=True)
````

````python
# Criar categorias ordenadas para sal√°rio anual
ordem_salario = ['menos que $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +', 'na', 'n√£o informado']
df['salario_anual'] = pd.Categorical(df['salario_anual'], categories=ordem_salario, ordered=True)
````

````python
# Verificar transforma√ß√µes
df[['escolaridade', 'salario_anual']].head()
````

### Agora vamos responder √†s perguntas estrat√©gicas:

### 1. Quais perfis movimentam mais valor?

````python
# Agrupar por diferentes caracter√≠sticas e ver valor m√©dio movimentado
def analisar_movimentacao(agrupamento):
    return df.groupby(agrupamento)['valor_transacoes_12m'].agg(['mean', 'median', 'count']).sort_values('mean', ascending=False)
````

````python
# Por faixa salarial
movimentacao_salario = analisar_movimentacao('salario_anual')
print("Movimenta√ß√£o por faixa salarial:")
print(movimentacao_salario)
````

````python
# Por tipo de cart√£o
movimentacao_cartao = analisar_movimentacao('tipo_cartao')
print("\nMovimenta√ß√£o por tipo de cart√£o:")
print(movimentacao_cartao)
````

````python
# Por escolaridade
movimentacao_escolaridade = analisar_movimentacao('escolaridade')
print("\nMovimenta√ß√£o por escolaridade:")
print(movimentacao_escolaridade)
````

---

### 2. Qual a rela√ß√£o entre escolaridade, renda e engajamento?

````python
# Criar gr√°fico de rela√ß√£o entre escolaridade e renda
plt.figure(figsize=(12, 6))
sns.boxplot(x='escolaridade', y='valor_transacoes_12m', hue='salario_anual', data=df)
plt.title('Rela√ß√£o entre Escolaridade, Renda e Valor Movimentado')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

Interpreta√ß√£o do gr√°fico: 
Podemos observar que, em geral, quanto maior a escolaridade e o sal√°rio, maior o valor movimentado em transa√ß√µes. Por exemplo, pessoas com p√≥s-gradua√ß√£o e sal√°rio anual mais alto tendem a ter as maiores medianas de transa√ß√µes.

````python

# Correla√ß√£o entre meses de relacionamento e valor transacionado
correlacao = df[['meses_de_relacionamento', 'valor_transacoes_12m', 'qtd_transacoes_12m']].corr()
print("\nCorrela√ß√£o entre tempo de relacionamento e engajamento:")
print(correlacao)
````

---

### 3. Quem s√£o os clientes mais inativos? Por qu√™?

````python
# Identificar clientes inativos (mais meses inativos)
clientes_inativos = df.sort_values('meses_inativo_12m', ascending=False).head(20)
````

````python
# Analisar caracter√≠sticas dos clientes inativos
print("Caracter√≠sticas dos clientes mais inativos:")
print(clientes_inativos[['idade', 'sexo', 'salario_anual', 'tipo_cartao', 'meses_inativo_12m', 'valor_transacoes_12m']].describe())
````

````python
# Verificar rela√ß√£o entre inatividade e outras vari√°veis
plt.figure(figsize=(10, 6))
sns.scatterplot(x='meses_inativo_12m', y='valor_transacoes_12m', hue='tipo_cartao', data=df)
plt.title('Rela√ß√£o entre Inatividade e Valor Transacionado')
plt.show()
````

Interpreta√ß√£o do gr√°fico:
Podemos perceber que a maioria dos clientes com mais meses inativos tende a movimentar menos dinheiro, especialmente os que usam cart√µes do tipo b√°sico. J√° os cart√µes do tipo premium aparecem mais entre os clientes com menor inatividade e valores maiores de transa√ß√µes.

---

### 4. H√° potencial para campanhas espec√≠ficas ou upgrades de cart√£o?

````python
# Identificar clientes com potencial para upgrade
# Crit√©rio: valor transacionado alto mas cart√£o b√°sico
potencial_upgrade = df[(df['tipo_cartao'].isin(['blue', 'silver'])) & 
                       (df['valor_transacoes_12m'] > df['valor_transacoes_12m'].quantile(0.75))]

print(f"\nClientes com potencial para upgrade de cart√£o: {len(potencial_upgrade)}")
````

````python
# Analisar caracter√≠sticas desses clientes
print(potencial_upgrade[['idade', 'sexo', 'salario_anual', 'tipo_cartao', 'valor_transacoes_12m']].describe())
````

````python
# Sugest√£o de campanhas espec√≠ficas por perfil
campanhas_especificas = df.groupby(['salario_anual', 'escolaridade', 'tipo_cartao'])['id'].count().unstack().fillna(0)
print("\nDistribui√ß√£o para campanhas espec√≠ficas:")
print(campanhas_especificas)
````

### **Visualiza√ß√£o dos Resultados e Conclus√µes**
Vamos criar visualiza√ß√µes para comunicar os insights:

````python
# Gr√°fico 1: Valor m√©dio transacionado por tipo de cart√£o e faixa salarial
plt.figure(figsize=(14, 7))
sns.barplot(x='salario_anual', y='valor_transacoes_12m', hue='tipo_cartao', data=df, ci=None)
plt.title('Valor M√©dio Transacionado por Tipo de Cart√£o e Faixa Salarial')
plt.ylabel('Valor M√©dio Transacionado (12 meses)')
plt.xlabel('Faixa Salarial')
plt.xticks(rotation=45)
plt.legend(title='Tipo de Cart√£o')
plt.tight_layout()
plt.show()
````

Interpreta√ß√£o do gr√°fico: O gr√°fico mostra que o valor m√©dio transacionado aumenta com a faixa salarial, especialmente para clientes com cart√µes premium. Por outro lado, clientes com cart√µes b√°sicos, mesmo em faixas salariais maiores, tendem a movimentar menos do que os que usam cart√µes mais sofisticados.

````python
# Gr√°fico 2: Distribui√ß√£o de inatividade por tipo de cart√£o
plt.figure(figsize=(10, 6))
sns.boxplot(x='tipo_cartao', y='meses_inativo_12m', data=df)
plt.title('Distribui√ß√£o de Meses Inativos por Tipo de Cart√£o')
plt.xlabel('Tipo de Cart√£o')
plt.ylabel('Meses Inativos (√∫ltimos 12 meses)')
plt.show()
````

Interpreta√ß√£o do gr√°fico: O gr√°fico sugere que usu√°rios de cart√µes premium (como platinum ou black) tendem a ficar menos tempo inativos, enquanto usu√°rios de cart√µes b√°sicos apresentam mais meses sem movimenta√ß√£o. Isso pode indicar maior engajamento de clientes com cart√µes de maior categoria.

````python
# Gr√°fico 3: Rela√ß√£o entre limite de cr√©dito e valor transacionado
plt.figure(figsize=(10, 6))
sns.scatterplot(x='limite_credito', y='valor_transacoes_12m', hue='salario_anual', data=df, alpha=0.6)
plt.title('Rela√ß√£o entre Limite de Cr√©dito e Valor Transacionado')
plt.xlabel('Limite de Cr√©dito')
plt.ylabel('Valor Transacionado (12 meses)')
plt.show()
````

Interpreta√ß√£o do gr√°fico: O gr√°fico mostra uma tend√™ncia positiva: clientes com limite de cr√©dito mais alto geralmente transacionam valores maiores, especialmente entre aqueles com sal√°rios mais altos. Isso indica que sal√°rio e limite influenciam o comportamento de uso do cart√£o.
